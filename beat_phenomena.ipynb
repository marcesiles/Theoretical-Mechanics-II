```python
import numpy as np
import matplotlib.pyplot as plt

# Definicin de parmetros
A_magnitude = 1      # Magnitud de A
A_phase = 0          # Fase de A
B_magnitude = 0.5    # Magnitud de B
B_phase = np.pi / 4  # Fase de B

omega = 2 * np.pi    # Frecuencia angular
k = 1                # Nmero de onda
Delta_omega = 0.5    # Cambio en frecuencia angular
Delta_k = 0.2        # Cambio en nmero de onda

# Espacio y tiempo
x = np.linspace(0, 120, 500)
t = 0  # Puedes variar el tiempo para ver diferentes instantes

# Clculo de la parte real de la superposicin
Psi_1 = A_magnitude * np.cos(omega * t - k * x + A_phase)
Psi_2 = B_magnitude * np.cos((omega + Delta_omega) * t - (k + Delta_k) * x + B_phase)
Psi_total = Psi_1 + Psi_2

# Graficar
plt.figure(figsize=(10, 6))
plt.plot(x, Psi_total, label='Re($\Psi(x,t)$)', color='blue')
plt.title('Parte Real de la Superposicin de Ondas')
plt.xlabel('Posicin (x)')
plt.ylabel('Re($\Psi(x,t)$)')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definicin de parmetros
A_magnitude = 1      # Magnitud de A
A_phase = 0          # Fase de A
B_magnitude = 0.5    # Magnitud de B
B_phase = np.pi / 4  # Fase de B

omega = 2 * np.pi    # Frecuencia angular
k = 1                # Nmero de onda
Delta_omega = 0.5    # Cambio en frecuencia angular
Delta_k = 0.2        # Cambio en nmero de onda

# Espacio y tiempo
x = np.linspace(0, 100, 500)
t = 0  # Puedes variar el tiempo para ver diferentes instantes

# Clculo de la superposicin
Psi_1 = A_magnitude * np.exp(1j * (omega * t - k * x + A_phase))
Psi_2 = B_magnitude * np.exp(1j * ((omega + Delta_omega) * t - (k + Delta_k) * x + B_phase))
Psi_total = Psi_1 + Psi_2

# Parte real y parte imaginaria
Psi_real = np.real(Psi_total)
Psi_imag = np.imag(Psi_total)

# Graficar
plt.figure(figsize=(12, 8))

# Grfica de la parte real
plt.plot(x, Psi_real, label='Re($\Psi(x,t)$)', color='blue')

# Grfica de la parte imaginaria
plt.plot(x, Psi_imag, label='Im($\Psi(x,t)$)', color='orange', linestyle='--')

plt.title('Superposicin de la Parte Real e Imaginaria de la Funcin de Onda')
plt.xlabel('Posicin (x)')
plt.ylabel('$\Psi(x,t)$')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid()
plt.legend()
plt.show()
```
